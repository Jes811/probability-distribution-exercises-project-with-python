# Probability distribution problems

## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment; for example, when you flip a coin, it can be either head or tails.

```python




from scipy.stats import norm

def dnorm(x, mean=0, sd=1):
    """
    Calculates the point estimate (probability density) of the normal distribution.
    Equivalent to R's dnorm().
    """
    result = norm.pdf(x, loc=mean, scale=sd)
    return result

def pnorm(q, mean=0, sd=1):
    """
    Calculates the cumulative probability of the normal distribution.
    Equivalent to R's pnorm().
    """
    result = norm.cdf(q, loc=mean, scale=sd)
    return result

def qnorm(p, mean=0, sd=1):
    """
    Calculates the quantile (inverse CDF) of the normal distribution.
    Equivalent to R's qnorm().
    """
    result = norm.ppf(p, loc=mean, scale=sd)
    return result

def rnorm(n, mean=0, sd=1):
    """
    Generates random variables from the normal distribution.
    Equivalent to R's rnorm().
    """
    result = norm.rvs(loc=mean, scale=sd, size=n)
    return result
```

```python
from scipy.stats import uniform

def dunif(x, min=0, max=1):
    """
    Calculates the probability density of the uniform distribution.
    Equivalent to R's dunif().
    """
    result = uniform.pdf(x, loc=min, scale=max - min)
    return result

def punif(q, min=0, max=1):
    """
    Calculates the cumulative probability of the uniform distribution.
    Equivalent to R's punif().
    """
    result = uniform.cdf(q, loc=min, scale=max - min)
    return result

def qunif(p, min=0, max=1):
    """
    Calculates the quantile (inverse CDF) of the uniform distribution.
    Equivalent to R's qunif().
    """
    result = uniform.ppf(p, loc=min, scale=max - min)
    return result

def runif(n, min=0, max=1):
    """
    Generates random variables from the uniform distribution.
    Equivalent to R's runif().
    """
    result = uniform.rvs(loc=min, scale=max - min, size=n)
    return result
```

```python
# Binomial(10, 0.2# Normal(0, 1) distribution example) distribution, where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial

import numpy as np
from scipy.stats import norm

# --- Normal distribution helper functions ---# Binomial(10, 0.2) distribution
# n = 10 is the number of trials
# p = 0.2 is the probability of success in a single trial

import numpy as np
from scipy.stats import binom

# Define helper functions
def dbinom(x, size, prob=0.5):
    """Calculates the point estimate (PMF) of the binomial distribution"""
    return binom.pmf(k=x, n=size, p=prob)

def pbinom(q, size, prob=0.5):
    """Calculates the cumulative probability of the binomial distribution"""
    return binom.cdf(k=q, n=size, p=prob)

def qbinom(p, size, prob=0.5):
    """Calculates the quantile (inverse CDF) of the binomial distribution"""
    return binom.ppf(q=p, n=size, p=prob)

def rbinom(n, size, prob=0.5):
    """Generates random samples from the binomial distribution"""
    return binom.rvs(n=size, p=prob, size=n)

# Example usage
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size=10, prob=0.2)}")
print(f"Probability that a Binomial(10, 0.2) takes a value less than 2: {pbinom(2, size=10, prob=0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size=10, prob=0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution:\n{rbinom(2000, size=10, prob=0.2)}")

def dnorm(x, mean=0, sd=1):
    """Probability density function"""
    return norm.pdf(x, loc=mean, scale=sd)

def pnorm(q, mean=0, sd=1):
    """Cumulative distribution function"""
    return norm.cdf(q, loc=mean, scale=sd)

def qnorm(p, mean=0, sd=1):
    """Quantile (inverse CDF)"""
    return norm.ppf(p, loc=mean, scale=sd)

def rnorm(n, mean=0, sd=1):
    """Random variable generation"""
    return norm.rvs(loc=mean, scale=sd, size=n)

# --- Run the example ---
np.random.seed(42)

print(f"Probability density that a Normal(0, 1) takes the value 0: {dnorm(0, mean=0, sd=1)}")
print(f"Probability that a Normal(0, 1) takes a value less than 1.0: {pnorm(1.0, mean=0, sd=1)}")
print(f"Which value of a Normal(0, 1) has a cumulative probability of 0.9? {qnorm(0.9, mean=0, sd=1)}")

import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
# Uniform(0, 1) distribution
# low = 0 is the minimum value
# high = 1 is the maximum value

import numpy as np
from scipy.stats import uniform

# Define helper functions
def dunif(x, low=0, high=1):
    """Calculates the probability density (PDF) of the uniform distribution"""
    return uniform.pdf(x, loc=low, scale=high - low)

def punif(q, low=0, high=1):
    """Calculates the cumulative probability (CDF) of the uniform distribution"""
    return uniform.cdf(q, loc=low, scale=high - low)

def qunif(p, low=0, high=1):
    """Calculates the quantile (inverse CDF) of the uniform distribution"""
    return uniform.ppf(p, loc=low, scale=high - low)

def runif(n, low=0, high=1):
    """Generates random samples from the uniform distribution"""
    return uniform.rvs(loc=low, scale=high - low, size=n)

# Example usage
np.random.seed(42)

print(f"Probability density that a Uniform(0, 1) takes the value 0.5: {dunif(0.5, low=0, high=1)}")
print(f"Probability that a Uniform(0, 1) takes a value less than 0.5: {punif(0.5, low=0, high=1)}")
print(f"Which value of a Uniform(0, 1) has a cumulative probability of 0.9? {qunif(0.9, low=0, high=1)}")
print(f"Generate 10 random values from a Uniform(0, 1) distribution:\n{runif(10, low=0, high=1)}")
```

```python

import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Normal distribution helper functions ---
def dnorm(x, mean=0, sd=1):
    """Probability density function"""
    return norm.pdf(x, loc=mean, scale=sd)

def pnorm(q, mean=0, sd=1):
    """Cumulative distribution function"""
    return norm.cdf(q, loc=mean, scale=sd)

def qnorm(p, mean=0, sd=1):
    """Quantile function"""
    return norm.ppf(p, loc=mean, scale=sd)

# --- Generate data ---
np.random.seed(42)
mean = 0
sd = 1
data = np.arange(-5, 5, 0.01)

# --- Compute values ---
pdf = dnorm(data, mean=mean, sd=sd)
cdf = pnorm(data, mean=mean, sd=sd)
ppf = qnorm(np.linspace(0.001, 0.999, len(data)), mean=mean, sd=sd)

# --- Plot ---
fig, axis = plt.subplots(3, 1, figsize=(6, 8))

axis[0].plot(data, pdf, "r-", label="PDF")
axis[1].plot(data, cdf, "b-", label="CDF")
axis[2].plot(np.linspace(0.001, 0.999, len(data)), ppf, "g-", label="PPF")

# Labels and titles
axis[0].set_title("Normal Distribution - PDF")
axis[1].set_title("Normal Distribution - CDF")
axis[2].set_title("Normal Distribution - PPF (Quantile Function)")

for ax in axis:
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()


np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# --- Uniform distribution helper functions ---
def dunif(x, low=0, high=1):
    """Probability density function"""
    return uniform.pdf(x, loc=low, scale=high - low)

def punif(q, low=0, high=1):
    """Cumulative distribution function"""
    return uniform.cdf(q, loc=low, scale=high - low)

def qunif(p, low=0, high=1):
    """Quantile function"""
    return uniform.ppf(p, loc=low, scale=high - low)

# --- Generate data ---
np.random.seed(42)
low = 0
high = 1
data = np.arange(-1, 2, 0.01)  # includes values outside [0,1] to show bounds

# --- Compute values ---
pdf = dunif(data, low=low, high=high)
cdf = punif(data, low=low, high=high)
ppf = qunif(np.linspace(0.001, 0.999, len(data)), low=low, high=high)

# --- Plot ---
fig, axis = plt.subplots(3, 1, figsize=(6, 8))

axis[0].plot(data, pdf, "r-", label="PDF")
axis[1].plot(data, cdf, "b-", label="CDF")
axis[2].plot(np.linspace(0.001, 0.999, len(data)), ppf, "g-", label="PPF")

# Titles and labels
axis[0].set_title("Uniform Distribution - PDF")
axis[1].set_title("Uniform Distribution - CDF")
axis[2].set_title("Uniform Distribution - PPF (Quantile Function)")

for ax in axis:
    ax.legend()
    ax.grid(True)

plt.tight_layout()
plt.show()
```
